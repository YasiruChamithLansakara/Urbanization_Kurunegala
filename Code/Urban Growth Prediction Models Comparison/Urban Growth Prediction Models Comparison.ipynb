## ü§ñ Models Compared

### 1. Prophet (Facebook Time Series)
- **Type**: Additive regression model
- **Features**: Yearly seasonality, multiplicative mode
- **Strengths**: Handles missing data, trend changes, outliers
- **Use Case**: Long-term forecasting with seasonal patterns
- **Best for**: Stable trends with clear seasonality

### 2. ARIMA (AutoRegressive Integrated Moving Average)
- **Type**: Statistical time series model
- **Parameters**: ARIMA(1,1,1) - order (p,d,q)
- **Strengths**: Classic econometric approach, interpretable
- **Use Case**: Short to medium-term forecasting
- **Best for**: Linear trends with stationary data

### 3. Random Forest Regressor
- **Type**: Machine learning ensemble
- **Parameters**: 300 trees, max depth 10
- **Features**: Month, Year, Quarter, Days Since Start
- **Strengths**: Captures non-linear patterns, robust to overfitting
- **Use Case**: Feature-based prediction with engineered variables
- **Best for**: Complex feature interactions

### 4. XGBoost (Extreme Gradient Boosting) ‚≠ê NEW
- **Type**: Advanced gradient boosting
- **Parameters**: 500 estimators, learning rate 0.05, max depth 6
- **Features**: Same as Random Forest + early stopping
- **Strengths**: Superior accuracy, handles non-linearity, efficient
- **Use Case**: High-performance prediction with tabular data
- **Best for**: Competitive accuracy on time series features

### 5. LSTM (Long Short-Term Memory) ‚≠ê NEW
- **Type**: Deep learning recurrent neural network
- **Architecture**: 2 LSTM layers (64 + 32 units) with dropout
- **Features**: Sequential patterns in time series
- **Strengths**: Captures long-term dependencies, non-linear patterns
- **Use Case**: Complex temporal dynamics and trend changes
- **Best for**: Non-linear growth with memory of past patterns

### 6. K-Means Clustering
- **Type**: Unsupervised learning
- **Parameters**: 3 clusters (growth phases)
- **Features**: Urban Area, Growth Rate, Time
- **Purpose**: Identify distinct urban development phases
- **Use Case**: Pattern recognition and regime change detection
- **Best for**: Understanding growth phase transitions

````python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from datetime import timedelta
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from prophet import Prophet
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.seasonal import seasonal_decompose
import xgboost as xgb
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
import warnings
warnings.filterwarnings('ignore')

print("="*70)
print("URBAN GROWTH PREDICTION MODELS - KURUNEGALA DISTRICT".center(70))
print("="*70)

# ===========================
# CONFIGURATION
# ===========================
BASE_DIR = r"D:\Satellite Image Processing\Urbanization_Kurunegala"
STATS_FILE = os.path.join(BASE_DIR, "Processed_Monthly", "Kurunegala_Urbanization_Stats.csv")
OUTPUT_DIR = os.path.join(BASE_DIR, "Urban_Growth_Predictions")
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ===========================
# LOAD DATA
# ===========================
def load_urbanization_data():
    """Load urbanization statistics from analysis"""
    
    if not os.path.exists(STATS_FILE):
        print(f"\n‚ùå ERROR: Statistics file not found: {STATS_FILE}")
        print("   Please run 'Urbanization Analysis.ipynb' Cell 3 first!")
        print("   This will generate the required CSV file.")
        return None
    
    print(f"\nüìÅ Loading data from: {STATS_FILE}")
    df = pd.read_csv(STATS_FILE, parse_dates=["Date"])
    df = df.sort_values("Date").reset_index(drop=True)
    
    print(f"\n‚úÖ Data loaded successfully")
    print(f"   Total records: {len(df)}")
    print(f"   Period: {df['Date'].min().date()} to {df['Date'].max().date()}")
    print(f"   Urban area range: {df['Urban_Area_km2'].min():.2f} - {df['Urban_Area_km2'].max():.2f} km¬≤")
    
    return df

print("\n" + "="*70)
print("LOADING URBANIZATION DATA...".center(70))
print("="*70)

df = load_urbanization_data()

if df is None:
    print("\n‚ö†Ô∏è  Cannot proceed without data.")
    print("   Run the urbanization analysis notebook first, then return here.")
else:
    # ===========================
    # DATA PREPARATION
    # ===========================
    print("\n" + "="*70)
    print("DATA SUMMARY".center(70))
    print("="*70)
    
    print(f"\nüìä Dataset Overview:")
    print("-" * 70)
    print(f"   Total observations: {len(df)}")
    print(f"   Years covered: {df['Date'].dt.year.nunique()}")
    print(f"   Average scenes per year: {len(df) / df['Date'].dt.year.nunique():.1f}")
    print(f"   Current urban area: {df['Urban_Area_km2'].iloc[-1]:.2f} km¬≤")
    print(f"   Total growth: {df['Urban_Area_km2'].iloc[-1] - df['Urban_Area_km2'].iloc[0]:.2f} km¬≤")
    print(f"   Growth percentage: {((df['Urban_Area_km2'].iloc[-1] - df['Urban_Area_km2'].iloc[0]) / df['Urban_Area_km2'].iloc[0]) * 100:.1f}%")
    
    # ===========================
    # TRAIN/TEST SPLIT
    # ===========================
    print("\n" + "="*70)
    print("PREPARING TRAIN/TEST SPLIT...".center(70))
    print("="*70)
    
    train_size = int(len(df) * 0.85)  # 85% for training, 15% for testing
    train_df = df.iloc[:train_size].copy()
    test_df = df.iloc[train_size:].copy()
    
    y_train = train_df["Urban_Area_km2"]
    y_test = test_df["Urban_Area_km2"]
    X_train_index = train_df["Date"]
    X_test_index = test_df["Date"]
    
    print(f"\nüìä Split Configuration:")
    print("-" * 70)
    print(f"   Training set: {len(train_df)} observations ({train_df['Date'].min().date()} to {train_df['Date'].max().date()})")
    print(f"   Test set: {len(test_df)} observations ({test_df['Date'].min().date()} to {test_df['Date'].max().date()})")
    print(f"   Train ratio: 85% | Test ratio: 15%")
    
    # ===========================
    # MODEL 1: PROPHET
    # ===========================
    print("\n" + "="*70)
    print("MODEL 1: PROPHET (Facebook Time Series)".center(70))
    print("="*70)
    
    prophet_df = train_df[["Date", "Urban_Area_km2"]].rename(columns={"Date": "ds", "Urban_Area_km2": "y"})
    
    print("\nüîÑ Training Prophet model...")
    print("   Configuration:")
    print("   ‚Ä¢ Yearly seasonality: Enabled")
    print("   ‚Ä¢ Changepoint prior scale: 0.3")
    print("   ‚Ä¢ Seasonality mode: Multiplicative")
    
    model_prophet = Prophet(
        yearly_seasonality=True,
        weekly_seasonality=False,
        daily_seasonality=False,
        changepoint_prior_scale=0.3,
        seasonality_mode='multiplicative'
    )
    model_prophet.fit(prophet_df)
    
    # Make predictions
    future = pd.DataFrame({'ds': test_df['Date'].values})
    forecast = model_prophet.predict(future)
    prophet_pred = forecast["yhat"].values
    
    # Calculate metrics
    prophet_mae = mean_absolute_error(y_test, prophet_pred)
    prophet_rmse = np.sqrt(mean_squared_error(y_test, prophet_pred))
    prophet_r2 = r2_score(y_test, prophet_pred)
    prophet_mape = np.mean(np.abs((y_test - prophet_pred) / y_test)) * 100
    
    print(f"\n‚úÖ Prophet Model Results:")
    print("-" * 70)
    print(f"   MAE:  {prophet_mae:.3f} km¬≤")
    print(f"   RMSE: {prophet_rmse:.3f} km¬≤")
    print(f"   R¬≤:   {prophet_r2:.3f}")
    print(f"   MAPE: {prophet_mape:.2f}%")
    
    # ===========================
    # MODEL 2: ARIMA
    # ===========================
    print("\n" + "="*70)
    print("MODEL 2: ARIMA (Statistical Time Series)".center(70))
    print("="*70)
    
    try:
        print("\nüîÑ Training ARIMA(1,1,1) model...")
        print("   Configuration: ARIMA(p=1, d=1, q=1)")
        
        arima_model = ARIMA(y_train, order=(1,1,1))
        arima_fit = arima_model.fit()
        arima_forecast = arima_fit.forecast(steps=len(y_test))
        
        arima_mae = mean_absolute_error(y_test, arima_forecast)
        arima_rmse = np.sqrt(mean_squared_error(y_test, arima_forecast))
        arima_r2 = r2_score(y_test, arima_forecast)
        arima_mape = np.mean(np.abs((y_test - arima_forecast) / y_test)) * 100
        
        print(f"\n‚úÖ ARIMA Model Results:")
        print("-" * 70)
        print(f"   MAE:  {arima_mae:.3f} km¬≤")
        print(f"   RMSE: {arima_rmse:.3f} km¬≤")
        print(f"   R¬≤:   {arima_r2:.3f}")
        print(f"   MAPE: {arima_mape:.2f}%")
        
    except Exception as e:
        print(f"\n‚ö†Ô∏è  ARIMA model failed: {str(e)}")
        print("   Using simple moving average as fallback...")
        
        window = min(12, len(y_train))
        arima_forecast = np.full(len(y_test), y_train.rolling(window=window).mean().iloc[-1])
        arima_mae = mean_absolute_error(y_test, arima_forecast)
        arima_rmse = np.sqrt(mean_squared_error(y_test, arima_forecast))
        arima_r2 = r2_score(y_test, arima_forecast)
        arima_mape = np.mean(np.abs((y_test - arima_forecast) / y_test)) * 100
        
        print(f"\n‚úÖ Moving Average Results (fallback):")
        print("-" * 70)
        print(f"   MAE:  {arima_mae:.3f} km¬≤")
        print(f"   RMSE: {arima_rmse:.3f} km¬≤")
        print(f"   R¬≤:   {arima_r2:.3f}")
        print(f"   MAPE: {arima_mape:.2f}%")
    
    # ===========================
    # MODEL 3: RANDOM FOREST
    # ===========================
    print("\n" + "="*70)
    print("MODEL 3: RANDOM FOREST (Machine Learning)".center(70))
    print("="*70)
    
    print("\nüîÑ Engineering features...")
    
    # Feature engineering
    df["Month"] = df["Date"].dt.month
    df["Year"] = df["Date"].dt.year
    df["Quarter"] = df["Date"].dt.quarter
    df["DayOfYear"] = df["Date"].dt.dayofyear
    df["Days_Since_Start"] = (df["Date"] - df["Date"].min()).dt.days
    
    feature_cols = ["Month", "Year", "Quarter", "Days_Since_Start"]
    
    X = df[feature_cols]
    y = df["Urban_Area_km2"]
    
    X_train_rf = X.iloc[:train_size]
    X_test_rf = X.iloc[train_size:]
    y_train_rf = y.iloc[:train_size]
    y_test_rf = y.iloc[train_size:]
    
    print(f"   Features: {', '.join(feature_cols)}")
    
    print("\nüîÑ Training Random Forest...")
    print("   Configuration:")
    print("   ‚Ä¢ Number of trees: 300")
    print("   ‚Ä¢ Max depth: 10")
    print("   ‚Ä¢ Random state: 42")
    
    rf = RandomForestRegressor(n_estimators=300, max_depth=10, random_state=42, n_jobs=-1)
    rf.fit(X_train_rf, y_train_rf)
    rf_pred = rf.predict(X_test_rf)
    
    rf_mae = mean_absolute_error(y_test_rf, rf_pred)
    rf_rmse = np.sqrt(mean_squared_error(y_test_rf, rf_pred))
    rf_r2 = r2_score(y_test_rf, rf_pred)
    rf_mape = np.mean(np.abs((y_test_rf - rf_pred) / y_test_rf)) * 100
    
    print(f"\n‚úÖ Random Forest Results:")
    print("-" * 70)
    print(f"   MAE:  {rf_mae:.3f} km¬≤")
    print(f"   RMSE: {rf_rmse:.3f} km¬≤")
    print(f"   R¬≤:   {rf_r2:.3f}")
    print(f"   MAPE: {rf_mape:.2f}%")
    
    # Feature importance
    feature_importance = pd.DataFrame({
        'Feature': feature_cols,
        'Importance': rf.feature_importances_
    }).sort_values('Importance', ascending=False)
    
    print(f"\nüìä Feature Importance:")
    print("-" * 70)
    for idx, row in feature_importance.iterrows():
        print(f"   {row['Feature']:20s}: {row['Importance']:.3f}")
    
    # ===========================
    # MODEL 4: XGBOOST
    # ===========================
    print("\n" + "="*70)
    print("MODEL 4: XGBOOST (Gradient Boosting)".center(70))
    print("="*70)
    
    print("\nüîÑ Training XGBoost model...")
    print("   Configuration:")
    print("   ‚Ä¢ Number of estimators: 500")
    print("   ‚Ä¢ Max depth: 6")
    print("   ‚Ä¢ Learning rate: 0.05")
    print("   ‚Ä¢ Early stopping: 50 rounds")
    
    xgb_model = xgb.XGBRegressor(
        n_estimators=500,
        max_depth=6,
        learning_rate=0.05,
        random_state=42,
        n_jobs=-1,
        early_stopping_rounds=50
    )
    
    xgb_model.fit(
        X_train_rf, y_train_rf,
        eval_set=[(X_test_rf, y_test_rf)],
        verbose=False
    )
    
    xgb_pred = xgb_model.predict(X_test_rf)
    
    xgb_mae = mean_absolute_error(y_test_rf, xgb_pred)
    xgb_rmse = np.sqrt(mean_squared_error(y_test_rf, xgb_pred))
    xgb_r2 = r2_score(y_test_rf, xgb_pred)
    xgb_mape = np.mean(np.abs((y_test_rf - xgb_pred) / y_test_rf)) * 100
    
    print(f"\n‚úÖ XGBoost Model Results:")
    print("-" * 70)
    print(f"   MAE:  {xgb_mae:.3f} km¬≤")
    print(f"   RMSE: {xgb_rmse:.3f} km¬≤")
    print(f"   R¬≤:   {xgb_r2:.3f}")
    print(f"   MAPE: {xgb_mape:.2f}%")
    print(f"   Best iteration: {xgb_model.best_iteration}")
    
    # ===========================
    # MODEL 5: LSTM
    # ===========================
    print("\n" + "="*70)
    print("MODEL 5: LSTM (Deep Learning)".center(70))
    print("="*70)
    
    print("\nüîÑ Preparing sequential data for LSTM...")
    
    # Create sequences for LSTM
    def create_sequences(data, seq_length):
        X, y = [], []
        for i in range(len(data) - seq_length):
            X.append(data[i:i+seq_length])
            y.append(data[i+seq_length])
        return np.array(X), np.array(y)
    
    # Scale data
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(df[["Urban_Area_km2"]].values)
    
    # Create sequences (lookback window of 12 observations)
    seq_length = min(12, len(train_df) // 3)
    X_lstm, y_lstm = create_sequences(scaled_data, seq_length)
    
    # Split into train/test
    train_size_lstm = len(train_df) - seq_length
    X_train_lstm = X_lstm[:train_size_lstm]
    y_train_lstm = y_lstm[:train_size_lstm]
    X_test_lstm = X_lstm[train_size_lstm:]
    y_test_lstm = y_lstm[train_size_lstm:]
    
    print(f"   Sequence length: {seq_length}")
    print(f"   Training sequences: {len(X_train_lstm)}")
    print(f"   Test sequences: {len(X_test_lstm)}")
    
    print("\nüîÑ Building LSTM network...")
    print("   Architecture:")
    print("   ‚Ä¢ Layer 1: LSTM(64 units) + Dropout(0.2)")
    print("   ‚Ä¢ Layer 2: LSTM(32 units) + Dropout(0.2)")
    print("   ‚Ä¢ Output: Dense(1)")
    print("   ‚Ä¢ Optimizer: Adam")
    print("   ‚Ä¢ Loss: MSE")
    
    # Build LSTM model
    lstm_model = Sequential([
        LSTM(64, return_sequences=True, input_shape=(seq_length, 1)),
        Dropout(0.2),
        LSTM(32, return_sequences=False),
        Dropout(0.2),
        Dense(1)
    ])
    
    lstm_model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    
    # Early stopping
    early_stop = EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True)
    
    print("\nüîÑ Training LSTM model...")
    
    # Train model
    history = lstm_model.fit(
        X_train_lstm, y_train_lstm,
        epochs=100,
        batch_size=8,
        validation_data=(X_test_lstm, y_test_lstm),
        callbacks=[early_stop],
        verbose=0
    )
    
    # Make predictions
    lstm_pred_scaled = lstm_model.predict(X_test_lstm, verbose=0)
    lstm_pred = scaler.inverse_transform(lstm_pred_scaled).flatten()
    y_test_lstm_original = scaler.inverse_transform(y_test_lstm.reshape(-1, 1)).flatten()
    
    lstm_mae = mean_absolute_error(y_test_lstm_original, lstm_pred)
    lstm_rmse = np.sqrt(mean_squared_error(y_test_lstm_original, lstm_pred))
    lstm_r2 = r2_score(y_test_lstm_original, lstm_pred)
    lstm_mape = np.mean(np.abs((y_test_lstm_original - lstm_pred) / y_test_lstm_original)) * 100
    
    print(f"\n‚úÖ LSTM Model Results:")
    print("-" * 70)
    print(f"   MAE:  {lstm_mae:.3f} km¬≤")
    print(f"   RMSE: {lstm_rmse:.3f} km¬≤")
    print(f"   R¬≤:   {lstm_r2:.3f}")
    print(f"   MAPE: {lstm_mape:.2f}%")
    print(f"   Training epochs: {len(history.history['loss'])}")
    print(f"   Final training loss: {history.history['loss'][-1]:.6f}")
    print(f"   Final validation loss: {history.history['val_loss'][-1]:.6f}")
    
    # ===========================
    # MODEL 6: K-MEANS CLUSTERING
    # ===========================
    print("\n" + "="*70)
    print("MODEL 6: K-MEANS CLUSTERING (Pattern Discovery)".center(70))
    print("="*70)
    
    print("\nüîÑ Performing clustering analysis...")
    print("   Number of clusters: 3")
    print("   Feature: Urban_Area_km2")
    
    X_scaled = StandardScaler().fit_transform(df[["Urban_Area_km2"]])
    kmeans = KMeans(n_clusters=3, n_init=10, random_state=42)
    df["Cluster"] = kmeans.fit_predict(X_scaled)
    
    print(f"\n‚úÖ Clustering complete")
    print(f"\nüìä Cluster Distribution:")
    print("-" * 70)
    for cluster_id in sorted(df['Cluster'].unique()):
        cluster_data = df[df['Cluster'] == cluster_id]
        print(f"   Cluster {cluster_id}: {len(cluster_data)} observations")
        print(f"      Urban area range: {cluster_data['Urban_Area_km2'].min():.2f} - {cluster_data['Urban_Area_km2'].max():.2f} km¬≤")
        print(f"      Time period: {cluster_data['Date'].min().date()} to {cluster_data['Date'].max().date()}")
        print()
    
    # Visualize clusters
    plt.figure(figsize=(14, 6))
    sns.scatterplot(x="Date", y="Urban_Area_km2", hue="Cluster", palette="Set2", data=df, s=80, alpha=0.8)
    plt.title("Temporal Clustering of Urban Growth - Kurunegala District", fontsize=14, fontweight='bold')
    plt.xlabel("Date", fontsize=12, fontweight='bold')
    plt.ylabel("Urban Area (km¬≤)", fontsize=12, fontweight='bold')
    plt.grid(True, alpha=0.3)
    plt.legend(title="Growth Phase", fontsize=10)
    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_DIR, "KMeans_Temporal_Clusters.png"), dpi=300, bbox_inches='tight')
    plt.show()
    
    # Save cluster data
    df.to_csv(os.path.join(OUTPUT_DIR, "Kurunegala_Clusters.csv"), index=False)
    print(f"üíæ Cluster data saved to: {OUTPUT_DIR}")
    
    # ===========================
    # MODEL COMPARISON
    # ===========================
    print("\n" + "="*70)
    print("MODEL PERFORMANCE COMPARISON".center(70))
    print("="*70)
    
    results = pd.DataFrame({
        "Model": ["Prophet", "ARIMA", "Random Forest", "XGBoost", "LSTM"],
        "MAE (km¬≤)": [prophet_mae, arima_mae, rf_mae, xgb_mae, lstm_mae],
        "RMSE (km¬≤)": [prophet_rmse, arima_rmse, rf_rmse, xgb_rmse, lstm_rmse],
        "R¬≤": [prophet_r2, arima_r2, rf_r2, xgb_r2, lstm_r2],
        "MAPE (%)": [prophet_mape, arima_mape, rf_mape, xgb_mape, lstm_mape]
    })
    
    print(f"\n{results.to_string(index=False)}")
    
    # Identify best model
    best_model_idx = results["RMSE (km¬≤)"].idxmin()
    best_model = results.loc[best_model_idx, "Model"]
    best_rmse = results.loc[best_model_idx, "RMSE (km¬≤)"]
    
    print(f"\nüèÜ BEST PERFORMING MODEL: {best_model}")
    print(f"   Lowest RMSE: {best_rmse:.3f} km¬≤")
    
    # Save comparison results
    results.to_csv(os.path.join(OUTPUT_DIR, "Model_Comparison_Results.csv"), index=False)
    print(f"\nüíæ Comparison results saved to: {OUTPUT_DIR}")
    
    # ===========================
    # VISUALIZATION: FORECAST COMPARISON
    # ===========================
    print("\n" + "="*70)
    print("GENERATING VISUALIZATIONS".center(70))
    print("="*70)
    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 14))
    
    # Plot 1: Full timeline
    ax1.plot(df["Date"], df["Urban_Area_km2"], 'ko-', label="Observed Data", linewidth=2, markersize=4, alpha=0.7)
    ax1.axvline(x=train_df['Date'].max(), color='red', linestyle='--', alpha=0.5, linewidth=2, label='Train/Test Split')
    ax1.plot(X_test_index, prophet_pred, 'g^-', label=f"Prophet (RMSE={prophet_rmse:.2f})", linewidth=2, markersize=5, alpha=0.8)
    ax1.plot(X_test_index, arima_forecast, 'bs-', label=f"ARIMA (RMSE={arima_rmse:.2f})", linewidth=2, markersize=5, alpha=0.8)
    ax1.plot(X_test_index, rf_pred, 'r*-', label=f"Random Forest (RMSE={rf_rmse:.2f})", linewidth=2, markersize=6, alpha=0.8)
    ax1.plot(X_test_index, xgb_pred, 'md-', label=f"XGBoost (RMSE={xgb_rmse:.2f})", linewidth=2, markersize=5, alpha=0.8)
    
    # LSTM predictions (aligned with test dates)
    lstm_test_dates = df['Date'].iloc[train_size_lstm + seq_length:train_size_lstm + seq_length + len(lstm_pred)]
    ax1.plot(lstm_test_dates, lstm_pred, 'c+-', label=f"LSTM (RMSE={lstm_rmse:.2f})", linewidth=2, markersize=7, alpha=0.8)
    
    ax1.set_title("Model Forecast Comparison - Kurunegala District", fontsize=15, fontweight='bold', pad=15)
    ax1.set_xlabel("Date", fontsize=12, fontweight='bold')
    ax1.set_ylabel("Urban Area (km¬≤)", fontsize=12, fontweight='bold')
    ax1.legend(fontsize=9, loc='upper left')
    ax1.grid(True, alpha=0.3)
    
    # Plot 2: Zoomed in on test period
    ax2.plot(X_test_index, y_test, 'ko-', label="Actual", linewidth=3, markersize=8)
    ax2.plot(X_test_index, prophet_pred, 'g^-', label=f"Prophet", linewidth=2, markersize=6, alpha=0.8)
    ax2.plot(X_test_index, arima_forecast, 'bs-', label=f"ARIMA", linewidth=2, markersize=6, alpha=0.8)
    ax2.plot(X_test_index, rf_pred, 'r*-', label=f"Random Forest", linewidth=2, markersize=7, alpha=0.8)
    ax2.plot(X_test_index, xgb_pred, 'md-', label=f"XGBoost", linewidth=2, markersize=6, alpha=0.8)
    ax2.plot(lstm_test_dates, lstm_pred, 'c+-', label=f"LSTM", linewidth=2, markersize=8, alpha=0.8)
    
    ax2.set_title("Test Period Comparison (Zoomed)", fontsize=14, fontweight='bold', pad=15)
    ax2.set_xlabel("Date", fontsize=12, fontweight='bold')
    ax2.set_ylabel("Urban Area (km¬≤)", fontsize=12, fontweight='bold')
    ax2.legend(fontsize=10)
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    
    # Save figure
    fig_path = os.path.join(OUTPUT_DIR, "Model_Forecast_Comparison.png")
    plt.savefig(fig_path, dpi=300, bbox_inches='tight')
    print(f"\n‚úÖ Visualizations saved to: {fig_path}")
    
    plt.show()
    
    print("\n" + "="*70)
    print("MODEL TRAINING AND COMPARISON COMPLETE".center(70))
    print("="*70)
````
