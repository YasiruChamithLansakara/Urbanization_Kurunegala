## Urbanization Analysis for Kurunegala District - Enhanced Urban Detection

This notebook performs detailed monthly urbanization analysis for **Kurunegala District**, Sri Lanka using Landsat satellite data from 2013-2025. 

### ðŸŽ¯ **Analysis Overview**

**Purpose**: Track urban growth patterns over time using satellite imagery to understand:
- Monthly urbanization trends in Kurunegala District
- Seasonal variations in urban detection
- Long-term growth patterns across 30 administrative divisions
- Urban change detection between time periods

**Data Source**: Landsat 8/9 Collection 2 Surface Reflectance data
- **Temporal Coverage**: 2013-2025 (13 years)
- **Spatial Resolution**: 30m pixels
- **Processing**: Monthly scenes processed individually
- **Area of Interest**: Kurunegala District (30 administrative divisions)

### ðŸ™ï¸ **Urban Classification Method**

**Multi-Index Approach for Clear Urban Identification:**
1. **NDBI** (Normalized Difference Built-up Index): Detects built-up areas
2. **NDVI** (Normalized Difference Vegetation Index): Separates vegetation from urban
3. **UI** (Urban Index): Combined indicator for robust urban detection
4. **MNDWI** (Modified Normalized Difference Water Index): Excludes water bodies

**Urban Pixel Criteria:**
- NDBI > 0.0 (Built-up indicator)
- NDVI < 0.3 (Low vegetation)
- UI > 0.0 (Urban Index positive)
- MNDWI < 0.0 (Not water)

### ðŸ“Š **Key Outputs**
1. **Monthly Statistics**: CSV file with urban area measurements for each scene
2. **Temporal Visualizations**: Trend charts showing urbanization over time
3. **Raster Products**: NDVI, NDBI, UI, and Urban classification maps for each date
4. **Change Maps**: Urban expansion/contraction between first and last scenes
5. **Seasonal Analysis**: Understanding seasonal variations in urban detection
6. **Urban Percentage**: Percentage of district area that is urbanized

```python
import os
import numpy as np
import rasterio
from rasterio.mask import mask
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
import glob
from datetime import datetime

# ===========================
# CONFIGURATION
# ===========================
BASE_DIR = ".."
SATELLITE_DATA_DIR = r"D:\Satellite Image Processing\Satellite Images"
AOI_PATH = os.path.join(BASE_DIR, "AOI/kurunegala_District_AOI.geojson")
OUTPUT_DIR = os.path.join(BASE_DIR, "Processed_Kurunegala_Urban")
os.makedirs(OUTPUT_DIR, exist_ok=True)

YEARS = range(2013, 2026)
BANDS = {
    "blue": "SR_B2.TIF",
    "green": "SR_B3.TIF", 
    "red": "SR_B4.TIF",
    "nir": "SR_B5.TIF",
    "swir1": "SR_B6.TIF",
    "swir2": "SR_B7.TIF",
}

# Urban classification thresholds for clear identification
URBAN_THRESHOLDS = {
    'ndbi_min': 0.0,      # Built-up areas have NDBI > 0
    'ndvi_max': 0.3,      # Urban areas have low vegetation (NDVI < 0.3)
    'ui_min': 0.0,        # Urban Index > 0
    'mndwi_max': 0.0      # Exclude water bodies (MNDWI < 0)
}

print("="*70)
print("URBANIZATION ANALYSIS FOR KURUNEGALA DISTRICT".center(70))
print("="*70)

# LOAD AOI
aoi = gpd.read_file(AOI_PATH)
print(f"\nâœ… AOI loaded: {len(aoi)} administrative divisions")
print(f"   CRS: {aoi.crs}")
print(f"   Total area: {aoi.geometry.area.sum() / 1e6:.2f} kmÂ²")

# ===========================
# HELPER FUNCTIONS
# ===========================

def clip_to_aoi(raster_path, aoi_geom, target_crs):
    """Clip raster to AOI boundary"""
    with rasterio.open(raster_path) as src:
        aoi_reprojected = aoi_geom.to_crs(target_crs)
        out_img, out_transform = mask(src, aoi_reprojected.geometry, crop=True)
        out_meta = src.meta.copy()
        out_meta.update({
            "driver": "GTiff",
            "height": out_img.shape[1],
            "width": out_img.shape[2],
            "transform": out_transform
        })
    return out_img[0].astype("float32"), out_meta

def extract_date_from_filename(filename):
    """Extract date from Landsat filename"""
    parts = filename.split('_')
    for part in parts:
        if len(part) == 8 and part.isdigit():
            year = int(part[:4])
            month = int(part[4:6])
            day = int(part[6:8])
            return datetime(year, month, day)
    return None

def find_all_landsat_scenes(year):
    """Find all Landsat scenes for a given year"""
    folder = os.path.join(SATELLITE_DATA_DIR, f"{year}")
    if not os.path.exists(folder):
        return []
    
    red_files = glob.glob(os.path.join(folder, f"*{BANDS['red']}"))
    scenes = []
    
    for red_file in red_files:
        filename = os.path.basename(red_file)
        scene_id = filename.replace(f"_{BANDS['red']}", "")
        date = extract_date_from_filename(filename)
        if date:
            scenes.append({
                'scene_id': scene_id,
                'date': date,
                'year': date.year,
                'month': date.month,
                'folder': folder
            })
    
    return sorted(scenes, key=lambda x: x['date'])

def find_scene_files(scene_info):
    """Find all band files for a specific scene"""
    scene_files = {}
    missing_bands = []
    
    for key, suffix in BANDS.items():
        file_path = os.path.join(scene_info['folder'], f"{scene_info['scene_id']}_{suffix}")
        if os.path.exists(file_path):
            scene_files[key] = file_path
        else:
            missing_bands.append(key)
    
    return scene_files, missing_bands

def calc_spectral_indices(blue, green, red, nir, swir1, swir2):
    """
    Calculate multiple spectral indices for robust urban detection
    
    Returns:
        ndvi: Normalized Difference Vegetation Index
        ndbi: Normalized Difference Built-up Index
        ui: Urban Index
        mndwi: Modified Normalized Difference Water Index
    """
    # Apply scale factor for Landsat Collection 2 Surface Reflectance
    blue = blue * 0.0000275 - 0.2
    green = green * 0.0000275 - 0.2
    red = red * 0.0000275 - 0.2
    nir = nir * 0.0000275 - 0.2
    swir1 = swir1 * 0.0000275 - 0.2
    swir2 = swir2 * 0.0000275 - 0.2
    
    # NDVI: Vegetation Index (vegetation = high, urban = low)
    ndvi = (nir - red) / (nir + red + 1e-10)
    
    # NDBI: Built-up Index (urban = high, vegetation = low)
    ndbi = (swir1 - nir) / (swir1 + nir + 1e-10)
    
    # UI: Urban Index (combines SWIR2 and NIR for urban areas)
    ui = (swir2 - nir) / (swir2 + nir + 1e-10)
    
    # MNDWI: Water Index (water = high, land = low)
    mndwi = (green - swir1) / (green + swir1 + 1e-10)
    
    return ndvi, ndbi, ui, mndwi

def classify_urban_enhanced(ndvi, ndbi, ui, mndwi):
    """
    Enhanced urban classification using multiple indices
    
    Urban criteria:
    - NDBI > 0.0 (built-up indicator)
    - NDVI < 0.3 (low vegetation)
    - UI > 0.0 (urban index positive)
    - MNDWI < 0.0 (exclude water)
    """
    urban_mask = (
        (ndbi > URBAN_THRESHOLDS['ndbi_min']) &
        (ndvi < URBAN_THRESHOLDS['ndvi_max']) &
        (ui > URBAN_THRESHOLDS['ui_min']) &
        (mndwi < URBAN_THRESHOLDS['mndwi_max'])
    )
    
    return np.where(urban_mask, 1, 0).astype("uint8")

# ===========================
# MAIN PROCESSING LOOP
# ===========================

stats = []
processed_scenes = []

print("\n" + "="*70)
print("SEARCHING FOR LANDSAT SCENES...".center(70))
print("="*70)

# Collect all scenes from all years
all_scenes = []
for year in YEARS:
    scenes = find_all_landsat_scenes(year)
    if scenes:
        all_scenes.extend(scenes)
        print(f"  {year}: {len(scenes)} scenes")

print(f"\nâœ… Found {len(all_scenes)} total scenes across {len([y for y in YEARS if find_all_landsat_scenes(y)])} years\n")

if len(all_scenes) == 0:
    print("âš ï¸  No Landsat data found!")
    print("    Please organize your data in year folders (2013, 2014, etc.)")
else:
    print("="*70)
    print("PROCESSING SCENES...".center(70))
    print("="*70 + "\n")
    
    for scene in tqdm(all_scenes, desc="Processing"):
        scene_files, missing_bands = find_scene_files(scene)
        
        if missing_bands:
            print(f"  âš ï¸  Missing bands for {scene['date'].strftime('%Y-%m-%d')}: {missing_bands}")
            continue
        
        try:
            # Get CRS from first band
            with rasterio.open(scene_files['red']) as src:
                raster_crs = src.crs
            
            # Load all bands
            band_data = {}
            for key, file_path in scene_files.items():
                arr, meta = clip_to_aoi(file_path, aoi, raster_crs)
                band_data[key] = arr

            # Calculate spectral indices
            blue, green, red, nir, swir1, swir2 = (
                band_data["blue"], band_data["green"], band_data["red"],
                band_data["nir"], band_data["swir1"], band_data["swir2"]
            )
            
            ndvi, ndbi, ui, mndwi = calc_spectral_indices(blue, green, red, nir, swir1, swir2)
            
            # Classify urban areas using enhanced method
            urban = classify_urban_enhanced(ndvi, ndbi, ui, mndwi)

            # Create filename with date
            date_str = scene['date'].strftime('%Y%m%d')
            
            # Save spectral indices
            meta.update(dtype=rasterio.float32, count=1)
            with rasterio.open(os.path.join(OUTPUT_DIR, f"NDVI_{date_str}.tif"), "w", **meta) as dst:
                dst.write(ndvi, 1)
            with rasterio.open(os.path.join(OUTPUT_DIR, f"NDBI_{date_str}.tif"), "w", **meta) as dst:
                dst.write(ndbi, 1)
            with rasterio.open(os.path.join(OUTPUT_DIR, f"UI_{date_str}.tif"), "w", **meta) as dst:
                dst.write(ui, 1)
            with rasterio.open(os.path.join(OUTPUT_DIR, f"MNDWI_{date_str}.tif"), "w", **meta) as dst:
                dst.write(mndwi, 1)

            # Save urban classification
            meta.update(dtype=rasterio.uint8)
            with rasterio.open(os.path.join(OUTPUT_DIR, f"Urban_{date_str}.tif"), "w", **meta) as dst:
                dst.write(urban, 1)

            # Calculate statistics
            total_pixels = urban.size
            valid_pixels = np.sum(~np.isnan(ndbi))
            urban_pixels = np.sum(urban == 1)
            urban_area_km2 = urban_pixels * 900 / 1e6  # 30m pixel = 900 mÂ²
            urban_percentage = (urban_pixels / valid_pixels) * 100 if valid_pixels > 0 else 0
            
            stats.append((
                scene['date'], 
                scene['year'], 
                scene['month'], 
                urban_pixels,
                valid_pixels,
                urban_area_km2,
                urban_percentage
            ))
            processed_scenes.append(scene)
            
            print(f"  âœ… {scene['date'].strftime('%Y-%m-%d')}: {urban_area_km2:.2f} kmÂ² ({urban_percentage:.2f}%)")
            
        except Exception as e:
            print(f"  âŒ Error processing {scene['date'].strftime('%Y-%m-%d')}: {str(e)}")
            continue

# ===========================
# RESULTS AND VISUALIZATION
# ===========================

if not stats:
    print("\n" + "="*70)
    print("NO DATA PROCESSED".center(70))
    print("="*70)
    print("\nPlease ensure you have Landsat data in year folders (2013-2025)")
else:
    print("\n" + "="*70)
    print("GENERATING RESULTS...".center(70))
    print("="*70 + "\n")
    
    # Create DataFrame
    df = pd.DataFrame(stats, columns=["Date", "Year", "Month", "Urban_Pixels", "Valid_Pixels", "Urban_Area_km2", "Urban_Percentage"])
    
    # Save statistics
    csv_path = os.path.join(OUTPUT_DIR, "Kurunegala_Urbanization_Stats.csv")
    df.to_csv(csv_path, index=False)
    print(f"âœ… Statistics saved: {csv_path}\n")
    
    # Display summary
    print("ðŸ“Š URBANIZATION SUMMARY FOR KURUNEGALA DISTRICT")
    print("-" * 70)
    print(f"   Analysis Period: {df['Date'].min().strftime('%Y-%m-%d')} to {df['Date'].max().strftime('%Y-%m-%d')}")
    print(f"   Total Scenes: {len(df)}")
    print(f"   Average Urban Area: {df['Urban_Area_km2'].mean():.2f} kmÂ²")
    print(f"   Minimum Urban Area: {df['Urban_Area_km2'].min():.2f} kmÂ² ({df['Urban_Percentage'].min():.2f}%)")
    print(f"   Maximum Urban Area: {df['Urban_Area_km2'].max():.2f} kmÂ² ({df['Urban_Percentage'].max():.2f}%)")
    print(f"   Urban Growth: {df['Urban_Area_km2'].max() - df['Urban_Area_km2'].min():.2f} kmÂ²")
    print(f"   Average Urbanization: {df['Urban_Percentage'].mean():.2f}% of district area")
    print("-" * 70 + "\n")
    
    print(df.head(10))
    
    # ===========================
    # VISUALIZATION 1: Monthly Trend
    # ===========================
    plt.figure(figsize=(16,8))
    plt.plot(df["Date"], df["Urban_Area_km2"], marker='o', color='#e74c3c', 
             linewidth=2, markersize=5, alpha=0.8, label='Urban Area')
    plt.fill_between(df["Date"], 0, df["Urban_Area_km2"], alpha=0.2, color='#e74c3c')
    plt.title("Monthly Urbanization Trend - Kurunegala District (2013-2025)", 
              fontsize=16, fontweight='bold', pad=20)
    plt.xlabel("Date", fontsize=13, fontweight='bold')
    plt.ylabel("Urban Area (kmÂ²)", fontsize=13, fontweight='bold')
    plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3, linestyle='--')
    plt.legend(fontsize=11)
    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_DIR, "Kurunegala_Monthly_Urbanization_Trend.png"), 
                dpi=300, bbox_inches='tight')
    plt.show()
    
    # ===========================
    # VISUALIZATION 2: Urban Percentage Over Time
    # ===========================
    plt.figure(figsize=(16,8))
    plt.plot(df["Date"], df["Urban_Percentage"], marker='s', color='#3498db', 
             linewidth=2, markersize=5, alpha=0.8, label='Urban %')
    plt.fill_between(df["Date"], 0, df["Urban_Percentage"], alpha=0.2, color='#3498db')
    plt.title("Urban Percentage of District Area - Kurunegala District", 
              fontsize=16, fontweight='bold', pad=20)
    plt.xlabel("Date", fontsize=13, fontweight='bold')
    plt.ylabel("Urban Percentage (%)", fontsize=13, fontweight='bold')
    plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3, linestyle='--')
    plt.legend(fontsize=11)
    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_DIR, "Kurunegala_Urban_Percentage_Trend.png"), 
                dpi=300, bbox_inches='tight')
    plt.show()

    # ===========================
    # VISUALIZATION 3: Yearly Statistics
    # ===========================
    yearly_stats = df.groupby('Year').agg({
        'Urban_Area_km2': ['mean', 'min', 'max', 'std'],
        'Urban_Percentage': 'mean'
    }).reset_index()
    yearly_stats.columns = ['Year', 'Mean_Urban_km2', 'Min_Urban_km2', 'Max_Urban_km2', 'Std_Urban_km2', 'Mean_Urban_Pct']
    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14,10))
    
    # Urban area with error bars
    ax1.errorbar(yearly_stats['Year'], yearly_stats['Mean_Urban_km2'], 
                yerr=yearly_stats['Std_Urban_km2'], marker='o', color='#2ecc71', 
                linewidth=2.5, markersize=8, capsize=6, capthick=2, label='Mean Â± Std')
    ax1.fill_between(yearly_stats['Year'], yearly_stats['Min_Urban_km2'], 
                     yearly_stats['Max_Urban_km2'], alpha=0.2, color='#27ae60', label='Min-Max Range')
    ax1.set_title("Yearly Urban Area Statistics - Kurunegala District", 
                  fontsize=14, fontweight='bold')
    ax1.set_ylabel("Urban Area (kmÂ²)", fontsize=12, fontweight='bold')
    ax1.grid(True, alpha=0.3, linestyle='--')
    ax1.legend(fontsize=10)
    
    # Urban percentage
    ax2.bar(yearly_stats['Year'], yearly_stats['Mean_Urban_Pct'], 
           color='#9b59b6', alpha=0.7, edgecolor='#8e44ad', linewidth=2)
    ax2.set_title("Average Urban Percentage by Year", fontsize=14, fontweight='bold')
    ax2.set_xlabel("Year", fontsize=12, fontweight='bold')
    ax2.set_ylabel("Urban Percentage (%)", fontsize=12, fontweight='bold')
    ax2.grid(True, alpha=0.3, linestyle='--', axis='y')
    
    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_DIR, "Kurunegala_Yearly_Statistics.png"), 
                dpi=300, bbox_inches='tight')
    plt.show()

    # ===========================
    # VISUALIZATION 4: Seasonal Analysis
    # ===========================
    # Map to Sri Lankan monsoon seasons
    df['Season'] = df['Month'].map({
        12: 'Northeast Monsoon', 1: 'Northeast Monsoon', 2: 'Northeast Monsoon',
        3: 'Inter-monsoon', 4: 'Inter-monsoon', 5: 'Southwest Monsoon',
        6: 'Southwest Monsoon', 7: 'Southwest Monsoon', 8: 'Southwest Monsoon',
        9: 'Southwest Monsoon', 10: 'Inter-monsoon', 11: 'Inter-monsoon'
    })
    
    seasonal_stats = df.groupby('Season')['Urban_Area_km2'].agg(['mean', 'std', 'count']).reset_index()
    
    plt.figure(figsize=(12,7))
    colors = ['#3498db', '#e74c3c', '#f39c12']
    bars = plt.bar(seasonal_stats['Season'], seasonal_stats['mean'], 
           yerr=seasonal_stats['std'], capsize=8, color=colors, 
           alpha=0.8, edgecolor='black', linewidth=1.5)
    
    # Add value labels on bars
    for i, bar in enumerate(bars):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                f'{height:.1f} kmÂ²\n(n={int(seasonal_stats["count"].iloc[i])})',
                ha='center', va='bottom', fontsize=10, fontweight='bold')
    
    plt.title("Seasonal Urban Area Analysis - Kurunegala District\n(Sri Lankan Monsoon Seasons)", 
              fontsize=14, fontweight='bold', pad=20)
    plt.xlabel("Season", fontsize=12, fontweight='bold')
    plt.ylabel("Mean Urban Area (kmÂ²)", fontsize=12, fontweight='bold')
    plt.grid(True, alpha=0.3, linestyle='--', axis='y')
    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_DIR, "Kurunegala_Seasonal_Analysis.png"), 
                dpi=300, bbox_inches='tight')
    plt.show()

    # ===========================
    # CHANGE DETECTION MAP
    # ===========================
    if len(processed_scenes) >= 2:
        try:
            print("\nðŸ—ºï¸  Creating urban change detection map...")
            
            first_scene = processed_scenes[0]
            last_scene = processed_scenes[-1]
            
            first_date = first_scene['date'].strftime('%Y%m%d')
            last_date = last_scene['date'].strftime('%Y%m%d')
            
            urban_first = rasterio.open(os.path.join(OUTPUT_DIR, f"Urban_{first_date}.tif")).read(1)
            urban_last = rasterio.open(os.path.join(OUTPUT_DIR, f"Urban_{last_date}.tif")).read(1)

            change = urban_last.astype(int) - urban_first.astype(int)
            
            # Save change map
            meta.update(dtype=rasterio.int8)
            change_path = os.path.join(OUTPUT_DIR, f"UrbanChange_{first_date}_to_{last_date}.tif")
            with rasterio.open(change_path, "w", **meta) as dst:
                dst.write(change.astype("int8"), 1)

            # Calculate change statistics
            new_urban = np.sum(change == 1)
            lost_urban = np.sum(change == -1)
            unchanged = np.sum(change == 0)
            
            new_urban_km2 = new_urban * 900 / 1e6
            lost_urban_km2 = lost_urban * 900 / 1e6
            
            print(f"\nâœ… Change Detection Results:")
            print(f"   Period: {first_scene['date'].strftime('%Y-%m-%d')} to {last_scene['date'].strftime('%Y-%m-%d')}")
            print(f"   New Urban Areas: {new_urban_km2:.2f} kmÂ²")
            print(f"   Lost Urban Areas: {lost_urban_km2:.2f} kmÂ²")
            print(f"   Net Change: {new_urban_km2 - lost_urban_km2:.2f} kmÂ²")
            
            # Visualize change
            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18,8))
            
            # First scene
            im1 = ax1.imshow(urban_first, cmap='Greys_r', vmin=0, vmax=1)
            ax1.set_title(f"Urban Areas - {first_scene['date'].strftime('%Y-%m-%d')}", 
                         fontsize=13, fontweight='bold')
            ax1.axis('off')
            plt.colorbar(im1, ax=ax1, label='Urban (1) / Non-urban (0)', shrink=0.7)
            
            # Change map
            cmap_change = plt.cm.colors.ListedColormap(['#3498db', '#ecf0f1', '#e74c3c'])
            im2 = ax2.imshow(change, cmap=cmap_change, vmin=-1, vmax=1)
            ax2.set_title(f"Urban Change Detection\n{first_scene['date'].strftime('%Y-%m-%d')} to {last_scene['date'].strftime('%Y-%m-%d')}", 
                         fontsize=13, fontweight='bold')
            ax2.axis('off')
            cbar = plt.colorbar(im2, ax=ax2, shrink=0.7, ticks=[-1, 0, 1])
            cbar.set_label('Urban Change', fontsize=11)
            cbar.ax.set_yticklabels(['Lost Urban', 'No Change', 'New Urban'])
            
            plt.tight_layout()
            plt.savefig(os.path.join(OUTPUT_DIR, "Kurunegala_Urban_Change_Map.png"), 
                       dpi=300, bbox_inches='tight')
            plt.show()
            
        except Exception as e:
            print(f"âŒ Could not create change map: {str(e)}")
    else:
        print("\nâš ï¸  Cannot create change map - need at least 2 scenes")

    print("\n" + "="*70)
    print("ANALYSIS COMPLETE!".center(70))
    print("="*70)
    print(f"\nâœ… Processed {len(stats)} scenes for Kurunegala District")
    print(f"ðŸ“ Results saved to: {OUTPUT_DIR}")
    print("\nOutput files:")
    print("  - Kurunegala_Urbanization_Stats.csv")
    print("  - Monthly trend charts (PNG)")
    print("  - Urban classification maps (GeoTIFF)")
    print("  - Spectral indices: NDVI, NDBI, UI, MNDWI (GeoTIFF)")
    print("="*70)
```
